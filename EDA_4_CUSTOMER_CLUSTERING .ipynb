### Data Cleaning and Preparation

Before diving into our analysis, we need to clean and organize the data to ensure it's accurate and reliable. The function `clean_and_assign_customer_ids` accomplishes this by performing a series of essential steps:

1. **Drop duplicated rows**:
   We start by dropping duplicated rows in the data.
   
2. **Remove Inconsistent Data**: 
   In this step, we eliminate rows where the `Quantity` is negative and the `Price` is greater than zero. These rows likely represent data entry errors since a negative quantity generally indicates a return, which should have a price of zero. By dropping these inconsistent rows, we improve the overall quality of our dataset.

3. **Assign Unique Customer IDs**:
   In some invoice groups, if there is a missing value (`NaN`) in the `Customer ID`, it means all `Customer IDs` in that group are missing. Since these groups make up a significant part of our data (about 20%), we cannot afford to drop them. Instead, we assign new unique `Customer IDs` starting from the highest current ID plus one. This ensures that each invoice group with missing `Customer IDs` gets a unique identifier, preserving the integrity and usefulness of the data.

4. **Remove Redundant Columns**:
   We drop the `Description` column from the DataFrame. Since each product is uniquely identified by its `StockCode`, the `Description` is redundant for our analysis. Removing it simplifies our dataset, making it easier to work with and reducing any potential noise.

5. **Flag Returned Goods**:
   To better understand customer behavior, we add a new column called `IsReturned`. This column indicates whether an item was returned (`Quantity < 0`) or not (`Quantity >= 0`). This additional information is valuable for analyzing return patterns and can help us tailor our strategies accordingly.

6. **Measure Performance and confirm the duplicates are dropped**:
   Finally, we measure how long the data cleaning process takes. By timing the execution, we can keep an eye on performance, especially when dealing with large datasets. Knowing the execution time helps us identify any bottlenecks and optimize the process if needed.

By following these steps, we ensure our dataset is clean, consistent, and ready for meaningful analysis. This clean dataset will serve as a solid foundation for gaining insights and making informed decisions.

```python
# Assuming 'new_df_cleaned2' is your original dataframe

# Create a copy for feature engineering
engineered_data = new_df_cleaned2.copy()

# Extracting date and time features
engineered_data['Year'] = engineered_data['InvoiceDate'].dt.year
engineered_data['Month'] = engineered_data['InvoiceDate'].dt.month
engineered_data['Day'] = engineered_data['InvoiceDate'].dt.day
engineered_data['Hour'] = engineered_data['InvoiceDate'].dt.hour
engineered_data['DayOfWeek'] = engineered_data['InvoiceDate'].dt.dayofweek
engineered_data['IsWeekend'] = engineered_data['DayOfWeek'].apply(lambda x: 1 if x >= 5 else 0)
engineered_data['IsBusinessHour'] = engineered_data['Hour'].apply(lambda x: 1 if 9 <= x <= 17 else 0)

# DayPeriod logic
def get_day_period(hour):
    if 6 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 18:
        return 'Afternoon'
    elif 18 <= hour < 24:
        return 'Evening'
    else:
        return 'Night'

engineered_data['DayPeriod'] = engineered_data['Hour'].apply(get_day_period)

# Season logic
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    elif month in [9, 10, 11]:
        return 'Fall'

engineered_data['Season'] = engineered_data['Month'].apply(get_season)

# Aggregation features
customer_agg = engineered_data.groupby('Customer ID').agg({
    'Quantity': 'sum',
    'Price': 'mean',
    'Invoice': 'count'
}).reset_index()
customer_agg.columns = ['Customer ID', 'TotalQuantity', 'AvgPrice', 'TransactionCount']

engineered_data = engineered_data.merge(customer_agg, on='Customer ID', how='left')

# Ratio and difference features
engineered_data['TotalAmount'] = engineered_data['Quantity'] * engineered_data['Price']
engineered_data['ReturnRate'] = engineered_data['IsReturned'] / engineered_data['TransactionCount']
engineered_data['AvgTransactionValue'] = engineered_data['TotalAmount'] / engineered_data['TransactionCount']

# Interaction features
engineered_data['Quantity_Price_Interaction'] = engineered_data['Quantity'] * engineered_data['Price']

# Month-End and Quarter-End features
engineered_data['MonthEnd'] = engineered_data['InvoiceDate'].dt.is_month_end.astype(int)
engineered_data['QuarterEnd'] = engineered_data['InvoiceDate'].dt.is_quarter_end.astype(int)

# Display the shape and sample of the engineered DataFrame
print(f"Engineered data shape: {engineered_data.shape}")
print("\nEngineered data sample:\n")
print(engineered_data.head())
```

```python
# Function to encode categorical columns using binary encoding
def binary_encode_column(column):
    """
    Manually encode a column into binary representation.
    """
    unique_values = column.unique()
    n_unique = len(unique_values)
    n_bits = int(np.ceil(np.log2(n_unique)))
    binary_map = {val: np.binary_repr(i, width=n_bits) for i, val in enumerate(unique_values)}
    binary_encoded = column.map(binary_map)
    binary_columns = np.array([list(map(int, list(val))) for val in binary_encoded])
    
    return binary_columns, [f"{column.name}_bin_{i}" for i in range(n_bits)]

def binary_encode_columns(df, columns):
    """
    Manually encode categorical columns using binary encoding.
    """
    df = df.copy()
    for column in columns:
        binary_columns, binary_column_names = binary_encode_column(df[column])
        binary_df = pd.DataFrame(binary_columns, columns=binary_column_names, index=df.index)
        df = pd.concat([df.drop(column, axis=1), binary_df], axis=1)
    return df

# Define categorical columns explicitly (excluding Customer ID)
categorical_columns = ['StockCode', 'Country', 'Invoice', 'IsReturned', 'Season', 'DayPeriod']

# Pipeline for binary encoding and scaling
def encode_and_scale(df, columns, exclude_columns):
    """
    Apply manual binary encoding and scaling to the dataframe, excluding specified columns.
    """
    df_encoded = binary_encode_columns(df, columns)
    
    df_excluded = df[exclude_columns]
    df_to_scale = df_encoded.drop(exclude_columns, axis=1)
    
    scaler = StandardScaler()
    scaled_data = scaler.fit_transform(df_to_scale)
    
    df_scaled = pd.DataFrame(scaled_data, columns=df_to_scale.columns)
    final_df = pd.concat([df_excluded.reset_index(drop=True), df_scaled.reset_index(drop=True)], axis=1)
    
    return final_df

# Specify columns to exclude from scaling (including Customer ID)
exclude_columns = ['Customer ID']

# Make a copy of engineered_data before dropping InvoiceDate
engineered_data_copy = engineered_data.copy()
engineered_data_copy = engineered_data_copy.drop(columns=['InvoiceDate'])

# Apply encoding and scaling, ensuring Customer ID is excluded
processed_df = encode_and_scale(engineered_data_copy, categorical_columns, exclude_columns)

# Display the shape and sample of the processed DataFrame
print(f"Processed data shape: {processed_df.shape}")
print("\nProcessed data sample:")
print(processed_df.head())
```
